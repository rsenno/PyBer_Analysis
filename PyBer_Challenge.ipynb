### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
# use the groupby() function to create a Series of data that has the type of city as the index, 
# then apply the count() method to the "ride_id" column.
total_rides_by_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_by_city_type
```

```python
# 2. Get the total drivers for each city type
# use the groupby() function to create a Series of data that has the type of city as the index, 
# then apply the sum() method to the "driver_count" column.
# total_drivers_by_city_type = pyber_data_df.groupby(["type"]).sum()["driver_count"]
# total_drivers_by_city_type
total_drivers_by_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_by_city_type
```

```python
#  3. Get the total amount of fares for each city type
# use the groupby() function to create a Series of data that has the type of city as the index, 
# then apply the sum() method to the "fare" column.
total_fares_by_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_by_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
# calculate the average fare per ride by city type by dividing the sum of all the fares by the total rides.
average_fare_per_ride = total_fares_by_city_type / total_rides_by_city_type
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
# calculate the average fare per driver by city type by dividing the sum of all the fares by the total drivers.
average_fare_per_driver = total_fares_by_city_type / total_drivers_by_city_type
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
# create a PyBer summary DataFrame with all the data gathered from Steps 1-5, using the column names:
# Total Rides, Total Drivers, Total Fares, Average Fare per Ride, Average Fare per Driver
# Note the index label will be "type".
pyber_summary_df = pd.DataFrame({
             "Total Rides": total_rides_by_city_type,
             "Total Drivers": total_drivers_by_city_type,
             "Total Fares": total_fares_by_city_type,
             "Average Fare per Ride": average_fare_per_ride,
             "Average Fare per Driver": average_fare_per_driver})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
#  remove the index name ("type") from the PyBer summary DataFrame.
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
# format the columns of the Pyber summary DataFrame to look like this:
# Total Rides, Total Drivers, Total Fares (with $, comma, and 2 decimal places), 
# Average Fare per Ride (with $, comma, and 2 decimal places), Average Fare per Driver (with $, comma, and 2 decimal places)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].astype('float64')
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].astype('float64')
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].astype('float64')

pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
# create a new DataFrame with multiple indices using the groupby() function on the 
# "type" and "date" columns of the pyber_data_df DataFrame, then apply the sum() method on the "fare" column to show 
# the total fare amount for each date.
pyber_data_by_type_and_date_df = pyber_data_df.groupby(["type","date"]).sum()["fare"]
pyber_data_by_type_and_date_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# use the provided code snippet to reset the index.
# df = df.reset_index()
pyber_data_by_type_and_date_df = pyber_data_by_type_and_date_df.reset_index()
pyber_data_by_type_and_date_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_data_by_type_and_date_pivot_df = pyber_data_by_type_and_date_df.pivot(index='date', columns='type', values='fare')
pyber_data_by_type_and_date_pivot_df
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
# pyber_data_by_type_and_date_pivot_df
pyber_pivot_Jan_Apr_df = pyber_data_by_type_and_date_pivot_df.loc['2019-01-01':'2019-04-29']
pyber_pivot_Jan_Apr_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
pyber_pivot_Jan_Apr_df.index = pd.to_datetime(pyber_pivot_Jan_Apr_df.index)
pyber_pivot_Jan_Apr_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pyber_pivot_Jan_Apr_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pyber_pivot_Jan_Apr_sum_by_week_df = pyber_pivot_Jan_Apr_df.resample('W').sum()
pyber_pivot_Jan_Apr_sum_by_week_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
# graph the resampled DataFrame from Step 7 (8 actually) using the object-oriented interface method and the df.plot() method, 
# as well as the Matplotlib "fivethirtyeight" graph style code snippet provided in the starter code. 
# Annotate the y-axis label and the title, then use the appropriate code to save the figure as PyBer_fare_summary.png 
# in your "analysis" folder.

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

pyber_pivot_Jan_Apr_sum_by_week_df.plot(figsize=(50,15), fontsize=40, linewidth=7.0)
plt.title("Total Fare by City Type", fontsize=50)
plt.ylabel("Fare ($USD)", fontsize=50)
plt.xlabel("")
plt.rcParams['legend.title_fontsize'] = '38'
plt.legend(loc='center', fontsize=38, title='type')
plt.tight_layout()
# Save Figure
plt.savefig("analysis/PyBer_fare_summary.png")
# The data points for the last week in April don't seem to match up with the picture in the Canvas Module Challenge page.
# Spoke with TA Gael Gatera who agreed that it's an anomaly in the provided CSV file, so ignore it. 
plt.show()
```
